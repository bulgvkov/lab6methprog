```python
class Array:
    """
    Массив данных
    """
    
    def __init__(self, arr):
        self._data = list(arr)
        """
        Конструктор массива 
        :arr: iterable
        """

    def __len__(self):
        return len(self._data)
        """
        Возвращает длину массива  
        :ret: lenght(array)
        """

    def __getitem__(self, item):
        return self._data[item]
        """
        Возвращает элемент массива по индексу 
        :item: индекс
        :ret: array[item]
        """

    def __delitem__(self, key):
        del self._data[key]
        """
        Удаляет элемент по индексу 
        :key: индекс
        :ret: none
        """

    def __add__(self, other):
        self._data += [other]
        return self
        """
        Добавляет элемент в массив 
        :other: элемент
        :ter: self
        """
    
    def returner(self):
        return self._data
```

```python
class SmallArray(Array):
    """
    Маленький массив
    """
    def __init__(self, arr):
        super().__init__(arr)
```

```python
class BigArray(Array):
    """
    Большой массив
    """
    def __init__(self, arr):
        super().__init__(arr)
        """
        Конструктор 
        :arr: iterable
        """

    def __getitem__(self, item):
        return bin(self._data[item])
        """
        Возвращает элементы в 16СС
        """
```

```python
class Set:
    """
    Класс множество. Внутри имеет различные типы данных в зависимости от размера(условие 4 варианта ч1).
    """
    maxs_len_small_array = 100 
    """
    Максимальный размер множества 1 типа
    """

    def __init__(self, array, cls=SmallArray):
        """
        Конструктор множества 
        :arr: iterable
        :cls: класс внутреннего устройства начального типа
        """
        self.cls = cls
        if len(array) > Set.maxs_len_small_array:
            self.cls = BigArray
        else:
            self._data = self.cls(array)

    def add(self, value):
        """
        Добавляет элемент в множество 
        :value: элемент 
        :ret: none
        """
        if self.cls != BigArray and (len(self._data) >= Set.maxs_len_small_array):
            self.cls = BigArray
            self._data = self.cls(self._data.returner())
        self._data += value

    def delete(self, index):
        """
        Удаляет элемент из множества по индексу 
        :index: индекс 
        :ret: none
        """
        if self.cls == BigArray and (len(self._data) == Set.maxs_len_small_array + 1):
            self.cls = SmallArray
            self._data = self.cls(self._data)
        del self._data[index]

    def get(self, index):
        """
        Возвращает элемент множества по индексу 
        :index: индекс 
        :ret: none
        """
        return self._data[index]
```

```python
class IterableSet(Set):
    """
    Множество с итератором
    """
    def __iter__(self):
        """
        Инициализирует цикл итерации 
        :ret: self
        """
        self.__n = -1
        return self

    def __next__(self):
        """
        Возвращает следующий элемент при итерации
        :ret: следующий  элемент
        """
        if self.__n == len(self._data) - 1:
            raise StopIteration
        self.__n += 1
        return self.get(self.__n)
```

```python
print("Маленький массив")
first = IterableSet([2, 4, 6, 8, 9])
for i in first:
    print(i)
print("Маленький массив размера 100, расширим его на 1")
second = IterableSet(range(100))
second.add(22)
for i, j in enumerate(second):
    if i < 13:
        print(j)
```
